```python
from sklearn.feature_extraction.text import CountVectorizer
cv =CountVectorizer(max_features=5000,stop_words='english')
# We have converted the tags values into vectors
vectors = cv.fit_transform(new_moviedf['Tags']).toarray()
```

```python
#Top 5k words used in the tags
cv.get_feature_names_out()
# Now we need to remove similar type of words like actions,action, actor,actors.... To make values unique
#and remove repetitve values we use technique of stemming
from nltk.stem.porter import PorterStemmer

stemming_w = PorterStemmer()
def stemming_f(text):
    z=[]
    #Take the string values in convert it into list
    for i in text.split():
        #iterate into the list and apply stemming mechanism to remove unnecessary repeative words actors/actor, 
        #activity/activities etc etc and append the data in a list
        z.append(stemming_w.stem(i))
#     print(" ".join(z))
    #return the appended data into a string 
    return " ".join(z)
        
```

```python
new_moviedf['Tags']=new_moviedf['Tags'].apply(stemming_f)
# stemming_f('action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. ingeniousfilmpartners twentiethcenturyfoxfilmcorporation duneentertainment lightstormentertainment samworthington zoesaldana sigourneyweaver jamescameron')
```
